```python
#This Python code defines a SpectralData class that is designed to store and organize spectral data
# Here's a breakdown of its purpose and functionality:
#The SpectralData class encapsulates: Wavenumbers (x-axis values, typically in cm⁻¹ for infrared/Raman spectra).
#Spectra (y-axis intensity values, often a 2D array for multiple samples).
#Groups (metadata/labels for classification, e.g., sample types or experimental conditions).
# Now, the function loads FTIR (Fourier-Transform Infrared) spectral data from CSV files, 
#organizes it by region codes, and returns a structured SpectralData object containing:
#Wavenumbers (x-axis) Spectral intensities (y-axis) Region-based group labels
```

```python
class SpectralData:
    def __init__(self, wavenumbers, spectra, groups):
        self.wavenumbers = wavenumbers
        self.spectra = spectra
        self.groups = groups
```

```python
def load_ftir_spectra(file_paths, region_codes=["KB", "NNB", "BB"]):
    """Load FTIR spectra from CSV files grouped by region codes"""
    all_spectra = []
    groups = []
    wavenumbers = None
    
    for file_path in file_paths:
        filename = os.path.basename(file_path)
        
        # Determine region from filename
        region = next((i for i, code in enumerate(region_codes) if code in filename), None)
        if region is None:
            print(f"Skipping {filename} - no region code matched")
            continue
            
        try:
            # Try reading CSV with different header options
            for header_option in [None, 0]:
                try:
                    data = pd.read_csv(file_path, header=header_option)
                    if data.shape[1] > 1:  # Has at least wavenumbers + 1 spectrum
                        break
                except:
                    continue
            
            if wavenumbers is None:
                wavenumbers = data.iloc[:, 0].values
            
            # Process all spectra columns
            for col in range(1, data.shape[1]):
                spectrum = data.iloc[:, col].values
                if not np.all(np.isnan(spectrum)) and not np.all(spectrum == 0):
                    all_spectra.append(spectrum)
                    groups.append(region)
                    
        except Exception as e:
            print(f"Error processing {filename}: {str(e)}")
            continue
    
    if not all_spectra:
        raise ValueError("No valid spectra loaded - check file formats")
    
    return SpectralData(
        wavenumbers=np.array(wavenumbers),
        spectra=np.array(all_spectra),
        groups=np.array(groups)
    )
```

```python
# ===== ENHANCED PCA WITH CLUSTERING ANALYSIS =====
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.spatial import ConvexHull
import matplotlib.patches as patches

try:
    print("\nPerforming PCA and clustering analysis...")
    
    # 1. Standardize and perform PCA
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(ftir_data.spectra)
    pca = PCA(n_components=5)
    principal_components = pca.fit_transform(X_scaled)
    
    # 2. Determine optimal clusters (elbow method)
    wcss = []
    max_clusters = min(7, len(X_scaled)-1)  # Maximum 7 clusters or n_samples-1
    for i in range(1, max_clusters+1):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
        kmeans.fit(X_scaled)
        wcss.append(kmeans.inertia_)
    
    # Find elbow point (optimal cluster count)
    diffs = np.diff(wcss)
    diff_ratios = diffs[:-1]/diffs[1:]
    optimal_clusters = np.argmax(diff_ratios) + 2  # +2 because of 0-index and diff
    
    # 3. Perform clustering with optimal cluster count
    kmeans = KMeans(n_clusters=optimal_clusters, random_state=42,  n_init=10)
    clusters = kmeans.fit_predict(X_scaled)
    
    # 4. Create enhanced visualization
    plt.figure(figsize=(12, 5))
    
    # Plot 1: Original region groups
    plt.subplot(1, 2, 1)
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # KB, NNB, BB colors
    for region_idx, region_name in enumerate(["KB", "NNB", "BB"]):
        mask = ftir_data.groups == region_idx
        if any(mask):
            plt.scatter(principal_components[mask, 0], 
                       principal_components[mask, 1],
                       color=colors[region_idx],
                       label=f"{region_name} (n={sum(mask)})",
                       alpha=0.6)
    
    plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
    plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
    plt.title('PCA by Beach Region')
    plt.legend()
    plt.grid(alpha=0.2)
    
    # Add label (a) to the first subplot
    plt.text(0.02, 0.98, '(a)', transform=plt.gca().transAxes, 
         fontsize=14, fontweight='bold', va='top')
    
    # Plot 2: Data-driven clusters
    plt.subplot(1, 2, 2)
    cluster_colors = plt.cm.tab10(np.linspace(0, 1, optimal_clusters))
    
    # Plot points and convex hulls for each cluster
    for cluster_idx in range(optimal_clusters):
        cluster_mask = clusters == cluster_idx
        plt.scatter(principal_components[cluster_mask, 0],
                   principal_components[cluster_mask, 1],
                   color=cluster_colors[cluster_idx],
                   label=f'Cluster {cluster_idx+1} (n={sum(cluster_mask)})',
                   alpha=0.6)
        
        # Add convex hull for cluster
        if sum(cluster_mask) > 2:  # Need at least 3 points for convex hull
            points = principal_components[cluster_mask]
            hull = ConvexHull(points)
            for simplex in hull.simplices:
                plt.plot(points[simplex, 0], points[simplex, 1], 
                        color=cluster_colors[cluster_idx], alpha=0.3)
    
    plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
    plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
    plt.title(f'PCA with {optimal_clusters} Data-Driven Clusters')
    plt.legend()
    plt.grid(alpha=0.2)
    
    plt.text(0.02, 0.98, '(b)', transform=plt.gca().transAxes, 
         fontsize=14, fontweight='bold', va='top')
    
    plt.tight_layout()
    
    # Save the figure as PDF
    output_pdf_path = os.path.join(data_dir, "PCA_Clustering_Analysis.pdf")
    plt.savefig(output_pdf_path, format='pdf', bbox_inches='tight', dpi=300)
    print(f"\nPCA plots saved as: {output_pdf_path}")
    plt.show()
    
    # 5. Print analysis results
    print(f"\nAnalysis Results:")
    print(f"- Optimal clusters detected: {optimal_clusters}")
    print(f"- PCA variance explained: PC1={pca.explained_variance_ratio_[0]*100:.1f}%, "
          f"PC2={pca.explained_variance_ratio_[1]*100:.1f}%")
    
    # Compare clusters with original regions
    if optimal_clusters > 0:
        print("\nCluster composition by beach region:")
        cluster_df = pd.DataFrame({
            'Region': ftir_data.groups,
            'Cluster': clusters
        })
        cross_tab = pd.crosstab(cluster_df['Region'], cluster_df['Cluster'])
        cross_tab.index = ["KB", "NNB", "BB"][:len(cross_tab)]
        print(cross_tab)

except Exception as e:
    print(f"\nError during clustering analysis: {str(e)}")

# ===== END OF ENHANCED ANALYSIS =====
```

```python
# ===== ENHANCED PREPROCESSING =====
from sklearn.preprocessing import StandardScaler
from sklearn.cross_decomposition import PLSRegression
from scipy.sparse import diags
from scipy.sparse.linalg import spsolve
import seaborn as sns

def baseline_als(y, lam=1e5, p=0.01, niter=10):
    """Asymmetric Least Squares baseline correction"""
    L = len(y)
    D = diags([1,-2,1], [0,-1,-2], shape=(L,L-2))
    w = np.ones(L)
    for _ in range(niter):
        W = diags(w, 0, shape=(L,L))
        Z = W + lam * D.dot(D.T)
        z = spsolve(Z, w*y)
        w = p * (y > z) + (1-p) * (y <= z)
    return z

try:
    print("\nPerforming enhanced preprocessing...")
    
    # 1. Load raw data (using your existing ftir_data)
    #X_raw = ftir_data.spectra
    X_transmittance = ftir_data.spectra
    
    
    wavenumbers = ftir_data.wavenumbers
    
    y = ftir_data.groups
    #print("groups:", y.shape)
    regions = ["KB", "NNB", "BB"]  # Your region names
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # KB, NNB, BB colors
    
    print("Converting transmittance to absorbance...")
    #X_absorbance = -np.log10(X_transmittance/100)
    X_absorbance = 1 - X_transmittance
    
    # 2. Baseline correction
    print("Applying baseline correction...")
    X_baseline = np.array([spectrum - baseline_als(spectrum) 
                          for spectrum in X_absorbance])
    
    # 3. Normalization (Area Under Curve)
    print("Applying area normalization...")
    X_norm = X_baseline / np.trapz(X_baseline, wavenumbers, axis=1)[:, np.newaxis]
    
    # 4. Standardization (mean-center + unit variance)
    #print("Standardizing data...")
    #X_scaled = StandardScaler().fit_transform(X_norm)
    
    # Create figure with 4 subplots
    fig, axes = plt.subplots(3, 1, figsize=(14, 10))
    axes = axes.ravel()
    processing_stages = [
        ("Original Spectra", X_absorbance),
        ("After Baseline Correction", X_baseline),
        ("After Normalization", X_norm)
        #("After Standardization", X_scaled)
    ]
    
    # Create proxy artists for legend (empty plots just for colors)
    legend_proxies = [plt.Line2D([0], [0], color=colors[i], lw=2, label=regions[i]) 
                     for i in range(3)]
    
    # Plot each processing stage
    for ax, (title, data) in zip(axes, processing_stages):
        for region_idx in range(3):
            region_mask = (y == region_idx)
            for spectrum in data[region_mask]:
                ax.plot(wavenumbers, spectrum, 
                       color=colors[region_idx],
                       alpha=0.4)
        ax.set_title(title)
        ax.set_xlabel("Wavenumber (cm$^{-1}$)")
        ax.grid(alpha=0.2)
    
    # Add single legend for all subplots
    fig.legend(handles=legend_proxies, 
               title="Region",
               loc='upper center',
               bbox_to_anchor=(0.5, 1.05),
               ncol=3)
    
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"\nPreprocessing error: {str(e)}")
# ===== END ENHANCED PREPROCESSING =====
```

```python
# ===== OPTIMIZED VIP ANALYSIS FOR YOUR SPECTRAL RANGE =====
from sklearn.preprocessing import StandardScaler
from sklearn.cross_decomposition import PLSRegression
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

plt.rcParams['text.usetex'] = True

try:
    print("\nPerforming optimized VIP analysis for 370-4000 cm$^{-1}$ range...")
    
    # Create a PDF file to save all plots
    output_pdf_path = os.path.join(data_dir, "VIP_Analysis_Results.pdf")
    
    # 1. Prepare data with full spectral range
    y = ftir_data.groups
    wavenumbers = ftir_data.wavenumbers
    group_names = ["KB", "NNB", "BB"]
    
    # 2. Verify spectral range
    print(f"\nSpectral range loaded: {wavenumbers[0]:.0f}-{wavenumbers[-1]:.0f} cm$^{-1}$")
    print(f"Total spectral points: {len(wavenumbers)}")
    
    # 3. Standardize full spectrum data
    X_scaled = StandardScaler().fit_transform(X_norm)
    
    # 4. Perform PLS-DA on full spectrum
    n_components = min(3, X_scaled.shape[1], X_scaled.shape[0]-1)
    pls = PLSRegression(n_components=n_components)
    pls.fit(X_scaled, y)
    
    # 5. Calculate VIP scores for full spectrum
    t = pls.x_scores_
    w = pls.x_weights_
    q = pls.y_loadings_
    p, h = w.shape
    vips = np.zeros((p,))
    s = np.diag(t.T @ t @ q.T @ q).reshape(h, -1)
    total_s = np.sum(s)
    
    for i in range(p):
        weight = np.array([(w[i,j] / np.linalg.norm(w[:,j]))**2 for j in range(h)])
        vips[i] = np.sqrt(p * np.sum(s * weight) / total_s)
        
    with PdfPages(output_pdf_path) as pdf:  # Properly indented block starts here
        # 6. Define key spectral regions of interest
        spectral_regions = {
            'Low-frequencies': (700, 1200),
            'Mid-frequencies': (1550, 1700), 
            'High-frequencies': (2900, 3500)
        }
        
        # 7. Create analysis for each region
        for region_name, (start_wn, end_wn) in spectral_regions.items():
            # Find indices within current region
            region_mask = (wavenumbers >= start_wn) & (wavenumbers <= end_wn)
            available_points = np.sum(region_mask)
            
            if available_points == 0:
                print(f"\nSkipping {region_name} - no data in {start_wn}-{end_wn} cm$^{-1}$ range")
                continue
            
            print(f"\nAnalyzing {region_name} region ({start_wn}-{end_wn} cm$^{-1}): {available_points} points available")
            
            region_vips = vips[region_mask]
            region_wn = wavenumbers[region_mask]
            region_X = X_scaled[:, region_mask]
            
            # Get top 4 bands in this region
            top_n = min(4, len(region_wn))
            top_idx = np.argsort(region_vips)[-top_n:][::-1]
            top_wn = region_wn[top_idx]
            top_vips = region_vips[top_idx]
            
            # Calculate group means for this region's top bands
            group_means = []
            for group in range(len(group_names)):
                group_mask = (y == group)
                if np.any(group_mask):
                    group_means.append(np.mean(region_X[group_mask][:, top_idx], axis=0))
                else:
                    group_means.append(np.zeros(top_n))
            
            # Create visualization
            plt.figure(figsize=(14, 6))
            
            # Heatmap with improved formatting
            ax = sns.heatmap(region_X[:, top_idx].T,
                            cmap='coolwarm',
                            yticklabels=top_wn.astype(int),
                            xticklabels=False,
                            center=0,
                            cbar_kws={'label': 'Standardized Intensity'})
            
            # Improved group labeling
            group_counts = np.bincount(y)
            cum_counts = np.cumsum(group_counts)
            
            # Add colored group backgrounds
            for i, (start, end) in enumerate(zip([0]+cum_counts.tolist()[:-1], cum_counts)):
                if group_counts[i] > 0:
                    ax.axvspan(start, end, facecolor=plt.cm.tab10(i), alpha=0.1)
                    ax.text((start+end)/2, -1.5, 
                           f"{group_names[i]}\n(n={group_counts[i]})",
                           ha='center', va='center',
                           fontsize=11,
                           color=plt.cm.tab10(i))
            
            # Add VIP scores with better positioning
            for i, wn in enumerate(top_wn):
                band_intensities = [mean[i] for mean in group_means]
                dominant_group = np.argmax(band_intensities)
                ax.text(region_X.shape[0] + 15, i + 0.5,
                       f"{top_vips[i]:.2f}\n({group_names[dominant_group]})",
                       ha='left', va='center',
                       fontsize=10,
                       color=plt.cm.tab10(dominant_group),
                       bbox=dict(facecolor='white', alpha=0.7, pad=1))
            
            plt.title(f'Top {top_n} Discriminative Bands in {region_name} Region\n{start_wn}-{end_wn} cm$^{-1}$', fontsize=14)
            plt.ylabel('Wavenumber (cm$^{-1}$)', fontsize=12)
            plt.xlabel('Samples', fontsize=12)
            plt.yticks(rotation=0)
            plt.tight_layout()
            
            pdf.savefig()  # Save to PDF
            plt.close()  # Close the figure
            
            
            plt.figure(figsize=(11, 8.5))  # Letter size
            plt.axis('off')
            
            # Print region-specific results
            print(f"Top bands in {region_name}:")
            for i, (wn, score) in enumerate(zip(top_wn, top_vips)):
                band_intensities = [mean[i] for mean in group_means]
                dominant_group = np.argmax(band_intensities)
                print(f"{i+1:2d}. {wn:.0f} cm$^{-1}$: VIP={score:.2f} (dominant in {group_names[dominant_group]})")
                print(f"    Intensity ratios: {dict(zip(group_names, [f'{x:.2f}' for x in band_intensities]))}")

except Exception as e:
    print(f"\nError during analysis: {str(e)}")
    if 'region_mask' in locals():
        print(f"Last region mask sum: {np.sum(region_mask)}")
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.metrics import (classification_report, 
                           confusion_matrix, 
                           precision_recall_fscore_support)
import matplotlib.pyplot as plt
import seaborn as sns

# =============================================
# 1. Data Preparation with Validation
# =============================================
X = np.asarray(ftir_data.spectra)
y = np.asarray(ftir_data.groups).ravel()  # Ensure y is flat 1D array

# Verify we have samples from all classes
unique_classes, class_counts = np.unique(y, return_counts=True)
print("Class distribution:")
for cls, count in zip(unique_classes, class_counts):
    print(f"{cls}: {count} samples")

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Standardize features
X_scaled = StandardScaler().fit_transform(X)

# =============================================
# 2. Robust PLS-DA with Cross-Validation
# =============================================
n_components = min(3, X_scaled.shape[1], X_scaled.shape[0]-1)
pls = PLSRegression(n_components=n_components)

# Ensure all classes are represented in each fold
cv = StratifiedKFold(n_splits=min(5, min(class_counts)), 
                    shuffle=True, 
                    random_state=42)

# Get predictions with error handling
try:
    y_pred = cross_val_predict(pls, X_scaled, y_encoded, cv=cv)
    y_pred_classes = np.argmax(y_pred, axis=1) if y_pred.ndim > 1 else (y_pred > 0.5).astype(int)
    y_pred_labels = le.inverse_transform(y_pred_classes)
except Exception as e:
    print(f"Prediction error: {str(e)}")
    raise

# =============================================
# 3. Enhanced Evaluation Metrics
# =============================================
def safe_classification_report(y_true, y_pred, target_names):
    """Handle cases where some classes have no predictions"""
    metrics = precision_recall_fscore_support(
        y_true, y_pred, 
        labels=np.arange(len(target_names)),
        zero_division=0  # Explicitly handle undefined metrics
    )
    
    report = {
        'precision': metrics[0],
        'recall': metrics[1],
        'f1-score': metrics[2],
        'support': metrics[3]
    }
    
    print("\nClassification Report:")
    print(f"{'Class':<10}{'Precision':<12}{'Recall':<10}{'F1-score':<10}{'Support':<10}")
    for i, name in enumerate(target_names):
        print(f"{name:<10}{report['precision'][i]:<12.2f}{report['recall'][i]:<10.2f}"
              f"{report['f1-score'][i]:<10.2f}{report['support'][i]:<10}")

safe_classification_report(y, y_pred_labels, le.classes_)

# Confusion matrix with all classes
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y, y_pred_labels, labels=le.classes_)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix (PLS-DA)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.tight_layout()
plt.show()

# =============================================
# 4. Final Model Training with VIP Scores
# =============================================
pls_final = PLSRegression(n_components=n_components)
pls_final.fit(X_scaled, y_encoded)

# Calculate VIP scores (same as before)
# ... [previous VIP calculation code] ...

# Plot VIP scores highlighting important regions
plt.figure(figsize=(12, 5))
plt.plot(ftir_data.wavenumbers, vips, 'k', linewidth=0.7)
plt.fill_between(ftir_data.wavenumbers, 0, vips, 
                where=(vips > 1), 
                color='red', alpha=0.3,
                label='VIP > 1 (Important)')
plt.xlabel('Wavenumber (cm-1)')
plt.ylabel('VIP Score')
plt.title('Variable Importance in Projection (VIP) Scores')
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
